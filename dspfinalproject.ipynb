```python
def clean_data(df_alcsmk):
    # Drop columns: 'INDICATOR_x', 'SUBJECT_x' and 8 other columns
    df_alcsmk = df_alcsmk.drop(columns=['INDICATOR_x', 'SUBJECT_x', 'MEASURE_x', 'FREQUENCY_x', 'Flag Codes_x', 'INDICATOR_y', 'SUBJECT_y', 'MEASURE_y', 'FREQUENCY_y', 'Flag Codes_y'])
    # Rename column 'TIME_x' to 'year_alc'
    df_alcsmk = df_alcsmk.rename(columns={'TIME_x': 'year_alc'})
    # Rename column 'Value_x' to 'value_alc'
    df_alcsmk = df_alcsmk.rename(columns={'Value_x': 'value_alc'})
    # Rename column 'TIME_y' to 'year_smk'
    df_alcsmk = df_alcsmk.rename(columns={'TIME_y': 'year_smk'})
    # Rename column 'Value_y' to 'value_smk'
    df_alcsmk = df_alcsmk.rename(columns={'Value_y': 'value_smk'})
    return df_alcsmk

df_alcsmk= clean_data(df_alcsmk.copy())
df_alcsmk.head()
```
