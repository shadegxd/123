$f(x_i + 2h) = \frac{(2h)^5 f^{(5)}(x_i)}{120}  $

$2f(x_i-h) = 2(\frac{(-h)^5 f^{(5)}(x_i)}{120}) \\$ 
$-f(x_i - 2h) = - \frac{(-2h^)5 f^{(5)}(x_i)}{120}\\$

Therefore, the difference is:  $$|f^{(3)} - (f^{(3)} + \frac{f^{(5)}(\xi)h^2}{2})| = \frac{f^{(5)}(\xi)h^2}{2}  $$
The error is : $\frac{f^{(5)}(\xi)h^2}{2}$ and grows in $O(h^2) $



We can conclude that $\triangle^3f_c$ is a second order approximation.

```python
import numpy as np 
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

def f(x):  #defining main function
    """The function f(x)""" 
    return x * np.exp(x+1)

def censc(x,h):   # defining the scheme
    """The Central Finite Difference scheme """
    return ((f(x + 2*h)) - (2 * (f(x + h))) + (2 * (f(x - h))) - (f(x - 2*h))) / (2 * h ** 3)

def f3(x):   #defining the third derivative
    """The third derivative of f(x)"""
    return 3 * np.exp(x+1) + np.exp(x+1) * x
def O(x): 
    """O(h^2)"""
    return x ** 2

hvalues= [10 ** -n for n in range(1,6)] # All the h values
errors = [] #list of errors
for i in hvalues:
    errors.append(abs(f3(0) - censc(0,i)))

plt.figure(figsize = (6,6)) 
plt.loglog(hvalues,errors,"--", label = "Error of $triangle^3f_c$")
plt.loglog(hvalues,[h ** 2 for h in hvalues], label = "$O(h^2)$") 
plt.legend(loc="best")
plt.xlabel("h values") 
plt.ylabel("errors") 
plt.show
```
