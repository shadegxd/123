```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

def c1_prime(t, c1, c2, L, V1):
    return - (L/V1) * c1 

def c2_prime(t, c1, c2, L, V2):
    return - (L/V2) * (c2 - c1)  

timespan = (0, 10)
c0 = [0.3, 0]  
L0 = 2.0
V11= 10.0
V22 = 5.0

sol = solve_ivp(fun=lambda t, y: [c1_prime(t, y[0], y[1], L0, V11),
                                  c2_prime(t, y[0], y[1], L0, V22)],
                t_span=timespan,
                y0=c0,
                t_eval=np.linspace(0, 10, 1000))

# Plot c1 and c2
plt.plot(sol.t, sol.y[0], label='Tank $1$: $C_1$', c="green")
plt.plot(sol.t, sol.y[1], label='Tank $2$: $C_2$', c="blue")
plt.xlabel('Time (minutes)')
plt.ylabel('Dye concentracion')
plt.legend()
plt.title("Amount of dye in two tanks")
plt.show()
```
